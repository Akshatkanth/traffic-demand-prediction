# Traffic Demand Prediction EDA

This notebook inspects the competition data, confirms the schema, and checks the key signals for modeling: timestamp structure, geohash cardinality, and missingness.

```python
from pathlib import Path

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

project_root = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
data_dir = project_root / 'data'
train = pd.read_csv(data_dir / 'train.csv')
test = pd.read_csv(data_dir / 'test.csv')
train.head()
```

```python
summary = pd.DataFrame({
    'train_shape': [train.shape],
    'test_shape': [test.shape],
    'train_missing': [train.isna().sum().sum()],
    'test_missing': [test.isna().sum().sum()],
})
summary
```

```python
missing = train.isna().mean().sort_values(ascending=False)
missing[missing > 0]
```
